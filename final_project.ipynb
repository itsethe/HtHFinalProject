```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt


```

```python
import csv
with open("Most_Streamed_Spotify_Songs_2024.csv", 'r') as file:
    df = csv.reader(file)


```

```python
top_spotify_songs = df.sort_values(by='Spotify Streams', ascending=False)
print(top_spotify_songs.head(10)[['Track', 'Spotify Streams']])
```

```python
spotify_data_sorted = df.sort_values(by='Spotify Streams', ascending=False)

# Get the top 10 songs
top_10 = spotify_data_sorted.head(10)

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_10['Track'], top_10['Spotify Streams'], color='skyblue')
plt.xlabel('Track')
plt.ylabel('Spotify Streams')
plt.title('Top 10 Most Streamed Spotify Songs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
num_explicit_songs = spotify_data_sorted['Explicit Track'].sum()

# Count the total number of songs
total_songs = spotify_data_sorted.shape[0]

# Calculate the percentage of explicit songs
percentage_explicit = (num_explicit_songs / total_songs) * 100

print(f"Percentage of explicit songs: {percentage_explicit:.2f}%")
```

```python
num_explicit_songs = spotify_data_sorted['Explicit Track'].sum()
num_non_explicit_songs = spotify_data_sorted.shape[0] - num_explicit_songs

# Create a list of labels and corresponding values
labels = ['Explicit', 'Non-Explicit']
sizes = [num_explicit_songs, num_non_explicit_songs]

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Explicit vs. Non-Explicit Songs')
plt.show()
```
